## Urbanization Analysis - Comprehensive Guide

This notebook performs detailed monthly urbanization analysis for Anuradhapura, Sri Lanka using Landsat satellite data from 2015-2025. 

### ðŸŽ¯ **Analysis Overview**

**Purpose**: Track urban growth patterns over time using satellite imagery to understand:
- Monthly urbanization trends
- Seasonal variations in urban detection
- Long-term growth patterns
- Urban change detection between time periods

**Data Source**: Landsat 8 Collection 2 Surface Reflectance data
- **Temporal Coverage**: 2015-2025 (11 years)
- **Spatial Resolution**: 30m pixels
- **Processing**: Monthly scenes processed individually
- **Area of Interest**: Anuradhapura district polygon

### ðŸ“Š **Key Outputs**
1. **Monthly Statistics**: CSV file with urban area measurements for each scene
2. **Temporal Visualizations**: Trend charts showing urbanization over time
3. **Raster Products**: NDVI, NDBI, and Urban classification maps for each date
4. **Change Maps**: Urban expansion/contraction between first and last scenes
5. **Seasonal Analysis**: Understanding seasonal variations in urban detection

```python
import os
import numpy as np
import rasterio
from rasterio.mask import mask
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import glob
from datetime import datetime

# CONFIGURATION
BASE_DIR = ".."
AOI_PATH = os.path.join(BASE_DIR, "AOI/Anuradhapura_District_AOI.geojson")
OUTPUT_DIR = os.path.join(BASE_DIR, "Processed_Monthly")
os.makedirs(OUTPUT_DIR, exist_ok=True)

YEARS = range(2015, 2026)
BANDS = {
    "blue": "SR_B2.TIF",
    "green": "SR_B3.TIF", 
    "red": "SR_B4.TIF",
    "nir": "SR_B5.TIF",
    "swir": "SR_B6.TIF",
}

# LOAD AOI
aoi = gpd.read_file(AOI_PATH)
print(f"AOI loaded: {len(aoi)} polygon(s) in {aoi.crs}")

# CLIP RASTER TO AOI
def clip_to_aoi(raster_path, aoi_geom, target_crs):
    with rasterio.open(raster_path) as src:
        aoi_reprojected = aoi_geom.to_crs(target_crs)
        out_img, out_transform = mask(src, aoi_reprojected.geometry, crop=True)
        out_meta = src.meta.copy()
        out_meta.update({
            "driver": "GTiff",
            "height": out_img.shape[1],
            "width": out_img.shape[2],
            "transform": out_transform
        })
    return out_img[0].astype("float32"), out_meta

# EXTRACT DATE FROM FILENAME
def extract_date_from_filename(filename):
    parts = filename.split('_')
    for part in parts:
        if len(part) == 8 and part.isdigit():
            year = int(part[:4])
            month = int(part[4:6])
            day = int(part[6:8])
            return datetime(year, month, day)
    return None

# FIND ALL LANDSAT SCENES FOR A GIVEN YEAR
def find_all_landsat_scenes(year):
    folder = os.path.join(BASE_DIR, f"Landsat_{year}")
    if not os.path.exists(folder):
        return []
    
    # Find all unique scene identifiers
    red_files = glob.glob(os.path.join(folder, f"*{BANDS['red']}"))
    scenes = []
    
    for red_file in red_files:
        filename = os.path.basename(red_file)
        scene_id = filename.replace(f"_{BANDS['red']}", "")
        date = extract_date_from_filename(filename)
        if date:
            scenes.append({
                'scene_id': scene_id,
                'date': date,
                'year': date.year,
                'month': date.month,
                'folder': folder
            })
    
    return sorted(scenes, key=lambda x: x['date'])

# FIND LANDSAT FILES FOR A SPECIFIC SCENE
def find_scene_files(scene_info):
    scene_files = {}
    missing_bands = []
    
    for key, suffix in BANDS.items():
        file_path = os.path.join(scene_info['folder'], f"{scene_info['scene_id']}_{suffix}")
        if os.path.exists(file_path):
            scene_files[key] = file_path
        else:
            missing_bands.append(key)
    
    return scene_files, missing_bands

# COMPUTE INDICES
def calc_indices(red, nir, swir):
    red = red * 0.0000275 - 0.2
    nir = nir * 0.0000275 - 0.2
    swir = swir * 0.0000275 - 0.2
    
    ndvi = (nir - red) / (nir + red + 1e-10)
    ndbi = (swir - nir) / (swir + nir + 1e-10)
    return ndvi, ndbi

# CLASSIFY URBAN AREAS
def classify_urban(ndbi, threshold=0.0):
    return np.where(ndbi > threshold, 1, 0).astype("uint8")

# MAIN PROCESSING LOOP
stats = []
processed_scenes = []

print("Finding all available Landsat scenes...")

# Collect all scenes from all years
all_scenes = []
for year in YEARS:
    scenes = find_all_landsat_scenes(year)
    all_scenes.extend(scenes)

print(f"Found {len(all_scenes)} total scenes")

for scene in tqdm(all_scenes, desc="Processing Landsat Scenes"):
    scene_files, missing_bands = find_scene_files(scene)
    
    if missing_bands:
        print(f"Missing bands for {scene['date'].strftime('%Y-%m-%d')}: {missing_bands}, skipping...")
        continue
    
    try:
        with rasterio.open(scene_files['red']) as src:
            raster_crs = src.crs
        
        band_data = {}
        for key, file_path in scene_files.items():
            arr, meta = clip_to_aoi(file_path, aoi, raster_crs)
            band_data[key] = arr

        red, nir, swir = band_data["red"], band_data["nir"], band_data["swir"]
        ndvi, ndbi = calc_indices(red, nir, swir)
        urban = classify_urban(ndbi)

        # Create filename with date
        date_str = scene['date'].strftime('%Y%m%d')
        
        meta.update(dtype=rasterio.float32, count=1)
        with rasterio.open(os.path.join(OUTPUT_DIR, f"NDVI_{date_str}.tif"), "w", **meta) as dst:
            dst.write(ndvi, 1)
        with rasterio.open(os.path.join(OUTPUT_DIR, f"NDBI_{date_str}.tif"), "w", **meta) as dst:
            dst.write(ndbi, 1)

        meta.update(dtype=rasterio.uint8)
        with rasterio.open(os.path.join(OUTPUT_DIR, f"Urban_{date_str}.tif"), "w", **meta) as dst:
            dst.write(urban, 1)

        valid_pixels = np.sum(~np.isnan(ndbi))
        urban_pixels = np.sum(urban == 1)
        urban_area_km2 = urban_pixels * 900 / 1e6
        
        stats.append((
            scene['date'], 
            scene['year'], 
            scene['month'], 
            urban_pixels, 
            urban_area_km2
        ))
        processed_scenes.append(scene)
        
        print(f"Processed {scene['date'].strftime('%Y-%m-%d')}: {urban_area_km2:.3f} kmÂ² urban area")
        
    except Exception as e:
        print(f"Error processing {scene['date'].strftime('%Y-%m-%d')}: {str(e)}")
        continue

if not stats:
    print("No data was successfully processed!")
else:
    # SAVE STATISTICS
    df = pd.DataFrame(stats, columns=["Date", "Year", "Month", "Urban_Pixels", "Urban_Area_km2"])
    csv_path = os.path.join(OUTPUT_DIR, "Monthly_Urbanization_Stats.csv")
    df.to_csv(csv_path, index=False)
    print(f"\nStatistics saved to {csv_path}\n")
    print(df.head(10))

    # PLOT MONTHLY URBANIZATION TREND
    plt.figure(figsize=(15,8))
    plt.plot(df["Date"], df["Urban_Area_km2"], marker='o', color='crimson', linewidth=1, markersize=4, alpha=0.7)
    plt.title("Monthly Urban Area Changes (Anuradhapura 2015-2025)", fontsize=14, fontweight='bold')
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Urban Area (kmÂ²)", fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "Monthly_Urbanization_Trend.png"), dpi=200, bbox_inches='tight')
    plt.show()

    # PLOT YEARLY AGGREGATED TREND
    yearly_stats = df.groupby('Year')['Urban_Area_km2'].agg(['mean', 'min', 'max', 'std']).reset_index()
    
    plt.figure(figsize=(12,6))
    plt.errorbar(yearly_stats['Year'], yearly_stats['mean'], 
                yerr=yearly_stats['std'], marker='o', color='darkred', 
                linewidth=2, markersize=6, capsize=5)
    plt.fill_between(yearly_stats['Year'], yearly_stats['min'], yearly_stats['max'], 
                    alpha=0.2, color='red', label='Min-Max Range')
    plt.title("Yearly Urban Area Statistics (Mean Â± Std)", fontsize=14, fontweight='bold')
    plt.xlabel("Year", fontsize=12)
    plt.ylabel("Urban Area (kmÂ²)", fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "Yearly_Urbanization_Stats.png"), dpi=200, bbox_inches='tight')
    plt.show()

    # SEASONAL ANALYSIS
    df['Season'] = df['Month'].map({
        12: 'Winter', 1: 'Winter', 2: 'Winter',
        3: 'Spring', 4: 'Spring', 5: 'Spring',
        6: 'Summer', 7: 'Summer', 8: 'Summer',
        9: 'Fall', 10: 'Fall', 11: 'Fall'
    })
    
    seasonal_stats = df.groupby('Season')['Urban_Area_km2'].agg(['mean', 'std']).reset_index()
    
    plt.figure(figsize=(10,6))
    plt.bar(seasonal_stats['Season'], seasonal_stats['mean'], 
           yerr=seasonal_stats['std'], capsize=5, color=['lightblue', 'lightgreen', 'orange', 'lightcoral'])
    plt.title("Seasonal Urban Area Analysis", fontsize=14, fontweight='bold')
    plt.xlabel("Season", fontsize=12)
    plt.ylabel("Mean Urban Area (kmÂ²)", fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "Seasonal_Urbanization.png"), dpi=200, bbox_inches='tight')
    plt.show()

    # CREATE CHANGE MAP (First vs Last available scene)
    if len(processed_scenes) >= 2:
        try:
            first_scene = processed_scenes[0]
            last_scene = processed_scenes[-1]
            
            first_date = first_scene['date'].strftime('%Y%m%d')
            last_date = last_scene['date'].strftime('%Y%m%d')
            
            urban_first = rasterio.open(os.path.join(OUTPUT_DIR, f"Urban_{first_date}.tif")).read(1)
            urban_last = rasterio.open(os.path.join(OUTPUT_DIR, f"Urban_{last_date}.tif")).read(1)

            change = urban_last - urban_first
            meta.update(dtype=rasterio.int8)
            with rasterio.open(os.path.join(OUTPUT_DIR, f"UrbanChange_{first_date}_{last_date}.tif"), "w", **meta) as dst:
                dst.write(change.astype("int8"), 1)

            print(f"Urban change map created: UrbanChange_{first_date}_{last_date}.tif")

            plt.figure(figsize=(10,8))
            plt.imshow(change, cmap='RdBu_r', vmin=-1, vmax=1)
            plt.title(f"Urban Change {first_scene['date'].strftime('%Y-%m-%d')} to {last_scene['date'].strftime('%Y-%m-%d')}\nRed = New Urban, Blue = Lost Urban", fontsize=14)
            plt.colorbar(label='Urban Change', shrink=0.8)
            plt.axis('off')
            plt.tight_layout()
            plt.savefig(os.path.join(OUTPUT_DIR, "Urban_Change_Map_Monthly.png"), dpi=200, bbox_inches='tight')
            plt.show()
            
        except Exception as e:
            print(f"Could not create change map: {str(e)}")
    else:
        print("Cannot create change map - insufficient data")

print(f"\nMonthly urbanization analysis complete! Processed {len(stats)} scenes.")
```
